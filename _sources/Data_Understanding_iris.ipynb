## Identifikasi Outlier
Outlier adalah data yang memiliki nilai jauh berbeda dibandingkan sebagian besar data lain.
Contoh: Pada dataset Iris, mayoritas SepalLength sekitar 5 cm, namun ada data yang bernilai 9 cm → ini disebut outlier.

### LOF

Apa itu LOF?
**LOF** adalah metode deteksi outlier berbasis **kepadatan (density)**.  
Data dianggap outlier jika berada di area yang **lebih jarang** dibanding tetangganya.

---

Cara Kerja
1. Cari **tetangga terdekat (kNN)** untuk tiap data.  
2. Hitung **kepadatan lokal** (seberapa rapat data di sekitar titik).  
3. Bandingkan kepadatan titik dengan tetangganya.  
   - Jika lebih rendah → **outlier**.  
   - Jika mirip → **normal**.  

---

Interpretasi
- **LOF ≈ 1** → normal (inlier).  
- **LOF > 1** → kemungkinan outlier.  
- **Semakin besar LOF**, semakin aneh datanya.  

---

Kelebihan & Kekurangan
Bisa deteksi **outlier lokal**.  
Butuh parameter **k** yang tepat dan agak lambat di dataset besar.

### KNN

K-Nearest Neighbors (KNN) untuk Deteksi Outlier

Apa itu KNN?
Metode **K-Nearest Neighbors (KNN)** digunakan untuk mendeteksi outlier dengan melihat **jarak** suatu data ke tetangga terdekatnya.  
Data dianggap outlier jika letaknya **jauh dari tetangga-tetangganya**.

---

Cara Kerja
1. Tentukan jumlah tetangga (**k**).  
2. Hitung jarak tiap data ke **k tetangga terdekat**.  
3. Jika rata-rata jarak data ke tetangga terlalu besar → data tersebut adalah **outlier**.  

---

Interpretasi
- **Jarak kecil** → data normal (inlier).  
- **Jarak besar** → data outlier.  

---

Kelebihan & Kekurangan
Mudah dipahami dan diimplementasikan.  
Cocok untuk data dengan distribusi sederhana.  
Sensitif terhadap pemilihan parameter **k**.  
Kurang efisien jika dataset sangat besar.

### ABOD

Angle-Based Outlier Detection (ABOD)

Apa itu ABOD?
**ABOD** adalah metode deteksi outlier berbasis **sudut (angle)** antar titik data.  
Ide dasarnya: data normal punya **sudut yang mirip** dengan banyak tetangganya, sedangkan outlier punya **sudut berbeda jauh**.

---

Cara Kerja
1. Hitung **sudut** antara vektor yang terbentuk dari suatu titik ke titik-titik lainnya.  
2. Bandingkan variasi sudut:  
   - Variasi sudut kecil → data **normal (inlier)**.  
   - Variasi sudut besar → data **outlier**.  

---

Interpretasi
- **Nilai ABOD tinggi** → normal.  
- **Nilai ABOD rendah** → kemungkinan outlier.  

---

Kelebihan & Kekurangan
Bagus untuk data berdimensi tinggi.  
Tidak hanya mengandalkan jarak, tapi juga pola sudut.  
Lebih lambat dihitung pada dataset besar.

## Kesimpulan

Berdasarkan analisis awal terhadap dataset Iris, dapat ditarik beberapa poin penting:

1. Dataset terdiri dari 150 sampel dengan distribusi yang seimbang, masing-masing spesies

   Hal ini terlihat dari nilai standar deviasi yang lebih tinggi pada Petal Length dan Petal Width.  

4. Fitur Sepal cenderung lebih stabil (rentang data lebih sempit dan standar deviasi lebih kecil).  
   Sebaliknya, fitur Petal lebih bervariasi sehingga berpotensi menjadi indikator utama dalam klasifikasi spesies.  

5. Secara keseluruhan, dataset ini sangat ideal untuk digunakan sebagai studi kasus dalam analisis data
   maupun pemodelan machine learning, karena memiliki data yang bersih, seimbang, dan jelas perbedaan antar fiturnya.
